```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a random correlation matrix
np.random.seed(0)
data = np.random.rand(10, 10)
corr = np.corrcoef(data)

# Create mask for lower triangle including diagonal
mask = np.tril(np.ones_like(corr, dtype=bool))

# Draw heatmap
sns.heatmap(corr, mask=mask, cmap='coolwarm', annot=True)
plt.title("Upper Triangle of Heatmap (Lower Triangle Masked)")
plt.show()
```

```python
import numpy as np
import pandas as pd

# Example item similarity matrix (5 items)
item_similarity = pd.DataFrame(
    [[1.0, 0.7, 0.2, 0.5, 0.3],
     [0.7, 1.0, 0.4, 0.6, 0.1],
     [0.2, 0.4, 1.0, 0.3, 0.8],
     [0.5, 0.6, 0.3, 1.0, 0.2],
     [0.3, 0.1, 0.8, 0.2, 1.0]],
    columns=['Item1', 'Item2', 'Item3', 'Item4', 'Item5'],
    index=['Item1', 'Item2', 'Item3', 'Item4', 'Item5']
)

# User's ratings for 3 items
user_ratings = {
    'Item1': 5.0,
    'Item2': 3.0,
    'Item3': 4.0
}

# Weighted sum of similarities
scores = {}
for item in item_similarity.columns:
    if item not in user_ratings:
        weighted_sum = sum(user_ratings[other] * item_similarity.loc[item, other] for other in user_ratings)
        similarity_sum = sum(item_similarity.loc[item, other] for other in user_ratings)
        scores[item] = weighted_sum / similarity_sum if similarity_sum != 0 else 0

# Top-5 recommendations
recommendations = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top-5 Recommendations:")
for item, score in recommendations:
    print(f"{item}: {score:.2f}")
```
